## Unsupervised neural networks
Instead of predicting a labeled value (supervised NNs), used for a variety of tasks

## Feature extraction
- Translates data from one set of features to a new set of different features
- New feature set is usually smaller and more suitable for further processing
- Example: taking series of pixels and translating them to lines, edges, or curves
- **Modeling pipeline:** process where extracted features are fed into another model for predictive purposes

## Restricted Boltzmann machines (RBM)
- Similar to single layer perceptrons but without a specified output to train on
- Works in two layers
    - 1st layer: inputs (visible nodes)
    - 2nd layer: hidden layer, new set of derived features
- Aims to extract patterns by finding common aggregations within the data

## Autoencoders
- Have an additional layer of output compared to RBMs
- Often defined by the number of features they output
    - Can have the same number of input & output features, used to better learn the existing features
    - Can have fewer output features, compressing the original data
    - Can have more output features, method to expand the data
- Same number of input/output features translated through a hidden layer with fewer features
    - Feature regeneration, improves robustness & groups redundancies
    - Output can be the same number as input, but in a better state for modeling
- Autoencoders are most often used early in the modeling pipeline for pre-training NNs
    - Uses hidden middle layer generated by the autoencoder as input for another NN layer

## Unsupervised NN usages
- Generally used in specific contexts
- Image recognition
    - RBM is valuable for this task
    - RBM is the only unsupervised NN in sklearn
    - For more, there is companion package skimage
